```r
library(glmnet)
```

glmnet does not take data frame as input, so we have to specify the x matrix and y vector.

```r
data = as.matrix(Boston[, -c(which(colnames(Boston)=='medv'))])
target = Boston$medv
lasso <- glmnet(x = data , y = target , alpha = 1)
#lambda = 0.5
coef(lasso,s=0.5)
```

```r
#lambda = 1
coef(lasso,s=1)
```

```r
plot(lasso, xvar = "lambda", label = TRUE)
```

```r
install.packages("plotmo")
library(plotmo) # for plot_glmnet
```

```r
plot_glmnet(lasso, label=TRUE)
```

```r
 # label the 5 biggest final coefs
plot_glmnet(lasso, label=5, xvar ="norm")
```

```r
# Using 5-fold cross-validation to search optimal lambda:
# use 5-fold cross validation to pick lambda
data = as.matrix(Boston[, -c(which(colnames(Boston)=='medv'))])
target = Boston$medv
cv_lasso <- cv.glmnet(x = data, y = target, alpha = 1, nfolds = 5)
plot(cv_lasso)
```

```r
# The best λ  (or s) is given by:
cv_lasso$lambda.min
```
