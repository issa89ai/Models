```r
library(tm)
library(NMF)
library(text)
```

```r
documents <- c(
  "I love reading about machine learning and artificial intelligence.",
  "Artificial intelligence is a fascinating field.",
  "I enjoy programming in Python.",
  "Python is great for data science and machine learning.",
  "Data science involves statistics and machine learning.",
  "The applications of AI are vast and varied.",
  "Learning data science can be fun and rewarding."
)
```

```r
# Create a Corpus
corpus <- Corpus(VectorSource(documents))

# Preprocess the text: convert to lower case, remove punctuation and stopwords
corpus <- tm_map(corpus, content_transformer(tolower))
corpus <- tm_map(corpus, removePunctuation)
corpus <- tm_map(corpus, removeWords, stopwords("en"))
corpus <- tm_map(corpus, stripWhitespace)

# Create Document-Term Matrix
dtm <- DocumentTermMatrix(corpus)
dtm_matrix <- as.matrix(dtm)
dtm_matrix <- dtm_matrix[rowSums(dtm_matrix) > 0, ]  # Remove empty rows
```

```r
# Set the number of topics
num_topics <- 2

# Perform NMF
nmf_result <- nmf(dtm_matrix, num_topics, nrun=10, .options='p')
```

```r
# Basis matrix (topics)
basis_matrix <- basis(nmf_result)
# Coefficient matrix (weights)
coeff_matrix <- coef(nmf_result)

# Print the matrices
print("Basis Matrix (Topics):")
print(basis_matrix)

print("Coefficient Matrix (Weights):")
print(coeff_matrix)
```

```r
# Display the top words for each topic
print("NMF Topics:")
for (topic_idx in 1:num_topics) {
  cat(sprintf("Topic %d:\n", topic_idx))
  top_words_indices <- order(basis_matrix[, topic_idx], decreasing = TRUE)[1:5]  # Get top 5 words
  top_words <- colnames(basis_matrix)[top_words_indices]
  cat(paste(top_words, collapse = ", "), "\n")
}
```
